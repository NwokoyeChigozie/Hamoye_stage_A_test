```python
import pandas as pd
url = "https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
fuel_ferc = pd.read_csv(url, error_bad_lines = False)
print(fuel_ferc)
```

## adding a list to another list using extend

```python
A = [1, 2, 3, 4, 5, 6] 
B = [13, 21, 34]

A.extend(B) 
print(A)
```

## sorting values in descending order based on values in the fuel_cost_per_unit_burned column

```python
fuel_ferc.sort_values(['fuel_cost_per_unit_burned'], ascending=[False])
```

## calculating average values based on the fuel_type_code_pudl values using groupby

```python
fuel_ferc.groupby(['fuel_type_code_pudl']).mean()
```

## getting the descriptive statistical data of all the columns in the dataframe

```python
fuel_ferc.describe(include = "all")
```

## calculating standard deviation and rounding to 2 decimal places

```python
stand = fuel_ferc.std(axis = 0, skipna = True)
print(round(stand,2))
```

## calculating skewness and rounding to 2 decimal places

```python
stand = fuel_ferc.skew(axis = 0, skipna = True)
print(round(stand,2))
```

## calculating kurtosis and rounding to 2 decimal places

```python
stand = fuel_ferc.kurtosis(axis = 0, skipna = True)
print(round(stand,2))
```

## checking for null values and calculating the total number of null values

```python
fuel_ferc.isnull().sum()
```
